```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
df = pd.merge(mouse_metadata, study_results, on='Mouse ID', how='outer')

# Display the data table for preview
df.head()
```

```python
# Checking the number of mice.
df['Mouse ID'].count()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
df[df.duplicated(subset=['Mouse ID', 'Timepoint'])][['Mouse ID', 'Timepoint']]
duplicated_mouse = df[df.duplicated(subset=['Mouse ID', 'Timepoint'])]['Mouse ID'].unique()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
df[(df['Mouse ID'].isin(duplicated_mouse))]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
df_cleaned = df[~(df['Mouse ID'].isin(duplicated_mouse))]
df_cleaned
```

```python
# Checking the number of mice in the clean DataFrame.
df_cleaned['Mouse ID'].count()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

capomulin_regimen = df_cleaned[(df_cleaned['Drug Regimen'] == 'Capomulin')]

capomulin_regimen = capomulin_regimen.groupby('Mouse ID').agg({'Weight (g)': 'mean', 'Tumor Volume (mm3)': 'mean'})

plt.figure(figsize=(9.5, 4))

plt.scatter(capomulin_regimen['Weight (g)'], capomulin_regimen['Tumor Volume (mm3)'])

plt.title('Capomulin Regimen - Average Tumor Volume (mm3) (y) vs. Mouse Weight (g) (x)')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.tight_layout()

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

# correlation coefficient
correlation = st.pearsonr(capomulin_regimen['Weight (g)'], capomulin_regimen['Tumor Volume (mm3)'])
print(f"The correlation between both factors is {round(correlation[0],2)}")
```

```python
# linear regression model 

x_values = capomulin_regimen['Weight (g)']
y_values = capomulin_regimen['Tumor Volume (mm3)']

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)

regress_values = x_values * slope + intercept

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.figure(figsize=(9.5, 4))

plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values, color='r')

# chart annotations
plt.annotate(f"The line equation is: {line_eq}",(20.5,36.5),fontsize=15)
plt.annotate(f"The r-squared value is: {round(rvalue**2,2)}",(20.5,35),fontsize=15)

plt.title('Capomulin Regimen - Average Tumor Volume (mm3) (y) vs. Mouse Weight (g) (x)')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.tight_layout()
plt.show()
```

### Observations
* There is a strong positive correlation (0.84) between mouse weight and tumor volume (mm3) across the Capomulin regimen - this means tumor volume (mm3) increases as mouse weight increases.
* We can also infer that tumor volume (mm3) significantly decreases as time passes under Capomulin regimen - this can be seen clearly in sample mouse s185as where tumor size dropped by almost 50% between timepoint 0 and 45.
* Mouse sex distribution was suprisingly even in the study, with males covering 51% and females 49% of the population.
* Capomulin was the most widely used regimen followed closely by Ramicane.
